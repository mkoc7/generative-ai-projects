## Generative AI Nedir?

Generative AI, verilen verilere benzer yeni içerikler üretebilen yapay zeka modellerini ifade eder. Bu modeller metin, görüntü, ses, video veya diğer veri türlerinde içerik oluşturabilir.

Generative AI'ın temel amacı, var olan verilerin dağılımını öğrenerek, bu dağılıma uygun yeni örnekler üretebilmektir.

```python
# Hugging Face Transformers kütüphanesini yükleme
!pip install transformers

# Basit bir metin üretme örneği
from transformers import pipeline

# Metin üretme modelini yükleme
generator = pipeline('text-generation', model='gpt2')

# Bir metin başlangıcı vererek devamını üretme
prompt = "Yapay zeka teknolojileri"
result = generator(prompt, max_length=50, num_return_sequences=1)

print(f"Prompt: {prompt}")
print(f"Üretilen metin: {result[0]['generated_text']}")
```
